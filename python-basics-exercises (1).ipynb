## Exercise 14: break and continue

1. Loop through numbers 1 to 20 and stop the loop when you reach 7.
2. Loop through numbers 1 to 10 but skip the number 5.

```python
# Your code here

for i in range(1,20):
    if i==7:
        break
    print(i)

print ("--------------------------------------------")

for i in range (1,10):
    if i==5:
        continue
    print(i)
```
