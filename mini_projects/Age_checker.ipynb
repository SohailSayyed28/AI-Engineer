```python
#Mini Project for an event i.e above 18
name = input("Enter your Name: ")
age_input = input("Enter your age: ")

try:
    # Attempt to convert to integer
    age = int(age_input)
    
    # Logic goes OUTSIDE the except block
    if 18 < age <= 64 :
        print(f"You may enter as your age is {age}")
    elif age >= 65:
        print(f"This event is not for you Veteran , but , it's your choice if you want to join")
    elif 13 <= age <= 17:
        print(f"You are not allowed teenager, look at your age. You're only {age}")
    else:
        print("Go home kiddo")

except ValueError:
    # This only runs if int(age_input) fails
    print("Please enter a valid number for your age.")
```
